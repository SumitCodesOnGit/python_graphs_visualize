```python
# matplotlib graph functions

import numpy as np
import matplotlib.pyplot as plt

# Line plot

def line_plot():
    x = np.linspace(0,10,100)
    y = np.sin(x)
    plt.plot(x,y)
    plt.title("Line Plot")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.show()



# Scatter Plot

def scatter_plot():
    x = np.random.rand(50)
    y = np.random.rand(50)
    plt.scatter(x, y, color = 'red')
    plt.title("Scatter Plot")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.grid(True)
    plt.show()


# Bar Plot

def bar_plot():
    categories = ['A','B','C']
    values = [3,7,5]
    plt.bar(categories,values)
    plt.xlabel("Categories")
    plt.ylabel("Values")
    plt.title("Bar Plot")
    plt.show()



# Histogram

def hist_plot():
    data = np.random.randn(1000)
    plt.hist(data, bins=50)
    plt.title("Histogram")
    plt.xlabel("Number")
    plt.ylabel("Distribution")
    plt.show()


# Pie chart

def pie_plot():
    labels = ['A', 'B', 'C', 'D']
    sizes = [30, 20, 40, 10]
    explode = [0, 0, 0.2, 0.1]
    plt.pie(sizes, labels=labels, explode=explode, autopct = '%1.1f%%')
    plt.title("Pie Chart")
    plt.show()


# Box Plot

def box_plot():
     data = [np.random.randn(1000) for _ in range(4)]
     plt.boxplot(data)
     plt.title("Box plot")
     plt.show()


# Fill in between plot

def fill_between():
    x = np.linspace(0,1,100)
    y1 = np.sin(2 * np.pi * x)
    y2 = 0.8 * np.sin(2 * np.pi * x)
    plt.fill_between(x, y1, y2, color='skyblue')
    plt.title("Fill Between")
    plt.show()


# Violin plot

def violin_plot():
    data = [np.random.normal(loc, 1.0, 100) for loc in [0, 1, 2, 3]]
    plt.violinplot(data, showmeans=True, showextrema=True, showmedians=True)
    plt.title("Violin Plot using Matplotlib")
    plt.xlabel("Group")
    plt.ylabel("Value")
    plt.grid(True)
    plt.show()


line_plot()
scatter_plot()
bar_plot()
hist_plot()
pie_plot()
box_plot()
fill_between()
violin_plot()
```

```python
# Seaborn graphs

import numpy as np
import seaborn as sns

df = sns.load_dataset("tips")
print(df.head())

# Line plot

x = np.linspace(0,10,100)
y = np.sin(x)
sns.lineplot( x=x, y=y)
plt.grid()
plt.title("Seaborn LinePlot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()


# Histogram
sns.histplot(df['total_bill'], kde=True)
plt.title("SNS Histplot")
plt.grid()
plt.xlabel("Total Bill")
plt.ylabel("Distribution")
plt.show()

# Scatter plot
sns.scatterplot(data=df, x = 'total_bill', y = 'tip')
plt.grid()
plt.title("Seaborn Scatter Plot")
plt.show()


# Bar Plot
sns.barplot(data=df, x='day', y='tip')
plt.title("Seaborn bar plot - average tip per day")
plt.show()


# Count plot
sns.countplot(data=df, x='day')
plt.title("Seaborn Count Plot for day")
plt.show()

sns.countplot(data=df, x= 'day', hue='sex')
plt.title("Seaborn Count plot for day split by sex")
plt.show()

# Pair plot
sns.pairplot(df, diag_kind = "kde")
plt.title("Seaborn Pair Plots")
plt.show()

# Heat Map
import pandas as pd
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Seaborn Heatmap")
plt.show()

# Swarm Plot
sns.swarmplot(data=df, x='day', y='total_bill')
plt.title("Swarm Plot")
plt.show()
```
